```python
import pandas as pd
```

```python
data = pd.read_csv('batch2_jobID_00B80TR.csv')
```

```python
data.head()
```

```python
#(Даша)середня зарплата всіх навчальних ступенів серед всіх індустрій ( можна помітити що чим вище освіта, тим вище заробіток)
def ch_col(df):
    return pd.Series(df.groupby("degree").salary.mean())
data.groupby("industry").apply(ch_col).loc[:, ["DOCTORAL", "MASTERS", "BACHELORS", "HIGH_SCHOOL", "NONE"]]
```

```python
#(Даша) кількість працівників, середній досвід працівників, середня зарплата в кожній компаніі
def ret_col(df):
    return pd.Series([df.jobType.count(), df.yearsExperience.mean(), df.salary.mean()], index=["Count of employee", "mean year exp", "mean salary"])
data.groupby("companyId").apply(ret_col).sort_values(by="mean salary", ascending=False)
```

```python
#(Даша)в якій індустрії найбільше менеджерів
data[data.jobType == "MANAGER"].industry.value_counts().idxmax()
```

```python
#(Cаша) Яка середня зарплата джуніора у сфері веб
round(data[(data.jobType == 'JUNIOR') & (data.industry == "WEB")].salary.sum() / len(data[(data.jobType == 'JUNIOR') & (data.industry == "WEB")].index))
```

```python
#(Саша) В якій сфері найбільша середня заробітня плата
data.groupby("industry").salary.mean().idxmax()
```

```python
# (Саша)Який відсоток працівників які займають високі посади не мають вищої освіти
a = len(data[((data.degree == 'HIGH_SCHOOL') | (data.degree == 'NONE')) & ((data.jobType == 'CFO') | (data.jobType == 'MANAGER') | (data.jobType == 'CEO') | (data.jobType == 'SENIOR') | (data.jobType == 'VICE_PRESIDENT') | (data.jobType == 'CTO'))].index)
b = len(data[(data.jobType == 'CFO') | (data.jobType == 'MANAGER') | (data.jobType == 'CEO') | (data.jobType == 'SENIOR') | (data.jobType == 'VICE_PRESIDENT') | (data.jobType == 'CTO')].index)
c = round(a/b *100 , 2)
c = str(c) + '%'
c
```

```python
#(Настя)Посортувати компанії в яких промисловість "здоров'я", від найменшої заробітньої плати до найбільшої
data.loc[data.industry == "HEALTH"].sort_values(by="salary")
```

```python
#(Настя) К-сть компаній, в яких відстані від метрополісу коливаються від 25-60
data.loc[(data.milesFromMetropolis <= 60) & (data.milesFromMetropolis >= 25)].companyId.count()
```

```python
#(Настя) Вивести максимальну зарплату серед тих у кого нема ступеня і кваліфікації
data.loc[(data.degree == "NONE") & (data.major == "NONE")].salary.max()
```

```python
#(Ярина)Знайти працівника з максимальним досвідом роботи
data.loc[data.yearsExperience.idxmax()]
```

```python
#(Ярина) знайти кількість працівників з досвідом роботи більше або рівним 10 років та зарплатою більше 130
data.loc[(data.yearsExperience >= 10) & (data.salary > 130)].jobType.count()
```

```python
#(Ярина) знайти скільки людей з нульовим досвідом та без вищої освіти заробляють більше середньої зарплати в кожній компанії
data.loc[(data.degree.isin(["NONE", "HIGH_SCHOOL"])) & (data.yearsExperience == 0) & (data.salary > data.salary.mean())].groupby("companyId").jobType.count()
```

```python
#(Анна-Марія) Скільки компаній розвиваються в сфері "Finance" 
data.loc[(data.industry == "FINANCE")].industry.count() 
```

```python
#(Анна-Марія) Яка мінімальна і максимальна зарплата в компаніях 
data.groupby('companyId').salary.agg([min,max]) 
```
